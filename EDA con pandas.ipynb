# ¿Qué es el EDA?
Eda es la sigla en inglés para **Exploratory Data Analysis** y consiste en una de las primeras tareas que tiene que desempeñar el Científico de Datos. Es cuando revisamos por primera vez los datos que nos llegan, por ejemplo un archivo CSV que nos entregan y deberemos intentar comprender **“¿de qué se trata?”, vislumbrar posibles patrones y reconociendo distribuciones estadísticas que puedan ser útiles en el futuro.**    
OJO!, **lo ideal es que tengamos un objetivo que nos hayan “adjuntado” con los datos**, que indique lo que se quiere conseguir a partir de esos datos. Por ejemplo, nos pasan un excel y nos dicen “Queremos predecir ventas a 30 días”, ó “Clasificar casos malignos/benignos de una enfermedad”, “Queremos identificar audiencias que van a realizar re-compra de un producto”, “queremos hacer pronóstico de fidelización de clientes/abandonos”, “Quiero detectar casos de fraude en mi sistema en tiempo real”.

# EDA deconstruido
Al llegar un archivo, lo primero que deberíamos hacer es intentar responder:

1. ¿Cuántos registros hay? ¿Son demasiado pocos? ¿Son muchos y no tenemos Capacidad (CPU+RAM) suficiente para procesarlo?
2. ¿Están todas las filas completas ó tenemos campos con valores nulos?
En caso que haya demasiados nulos: ¿Queda el resto de información inútil?
3. ¿Que datos son discretos y cuales continuos? * tipos de datos 
Muchas veces sirve obtener el tipo de datos: texto, int, double, float
4. tipo de problema: 
  
  4.1 Si es un problema de tipo supervisado:
¿Cuál es la columna de “salida”? ¿binaria, multiclase?
¿Esta balanceado el conjunto salida?

  4.2 ¿Cuales parecen ser features importantes? ¿Cuales podemos descartar?
¿Siguen alguna distribución?
  4.3 ¿Hay correlación entre features (características)?
En problemas de NLP es frecuente que existan categorías repetidas ó mal tipeadas, ó con mayusculas/minúsculas, singular y plural, por ejemplo “Abogado” y “Abogadas”, “avogado” pertenecerían todos a un mismo conjunto.
  4.4 ¿Estamos ante un problema dependiente del tiempo? Es decir un TimeSeries.
Si fuera un problema de Visión Artificial: ¿Tenemos suficientes muestras de cada clase y variedad, para poder hacer generalizar un modelo de Machine Learning?
5. ¿Cuales son los Outliers? (unos pocos datos aislados que difieren drásticamente del resto y “contaminan” ó desvían las distribuciones)
Podemos eliminarlos? es importante conservarlos?
son errores de carga o son reales?
6. ¿Tenemos posible sesgo de datos? (por ejemplo perjudicar a clases minoritarias por no incluirlas y que el modelo de ML discrimine)
Puede ocurrir que tengamos set de datos incompletos y debamos pedir a nuestro cliente/proveedor ó interesado que nos brinde mayor información de los campos, que aporte más conocimiento ó que corrija campos.

# Tipos de datos Estadísticos: Discretos, continuos, categóricos y ordinales
[Tipos de datos Estadísticos: Discretos, continuos, categóricos y ordinales](https://https://www.tecnologias-informacion.com/tipos-estadisticos.html#:~:text=Los%20datos%20discretos%20son%20un,lo%20general%2C%20implica%20n%C3%BAmeros%20enteros.&text=Por%20ejemplo%2C%20podr%C3%ADamos%20realizar%20un,el%20interior%20(datos%20discretos).)

Al trabajar con estadísticas, es importante reconocer los diferentes tipos de datos: numéricos (discretos y continuos), categóricos y ordinales.
La mayoría de los datos caen en uno de dos grupos: numéricos o categóricos.

**En términos generales, cuando mides algo y le das un valor numérico, creas datos cuantitativos. Cuando clasifica o juzga algo, crea datos cualitativos.** 
En el nivel más alto, existen dos tipos de datos: 
## 1 cuantitativo 
La información cuantitativa trata de números y cosas que puede medir objetivamente: dimensiones como altura, ancho y longitud. Temperatura y humedad.
Como regla general, los recuentos son discretos y las mediciones son continuas.
### 1.1 datos discretos
Los datos discretos son un conteo que no se puede hacer más preciso. Por lo general, implica números enteros. Por ejemplo,el numero de hijos, es un entero, no puedes tener 1.5 hijos. 
### 1.2 Los datos continuos
Los datos continuos, por otro lado, podrían dividirse y reducirse a niveles cada vez más finos. Por ejemplo, puede medir la altura de sus hijos en escalas progresivamente más precisas (metros, centímetros, milímetros y más).

Algunos análisis usan datos cuantitativos continuos y discretos al mismo tiempo.

Por ejemplo, podríamos realizar un análisis de regresión para ver si el peso de los manís (datos continuos) está correlacionado con el número de manís en el interior (datos discretos).

**También puede que nos pasen múltiples fuentes de datos, por ejemplo un csv, un excel y el acceso a una base de datos. Entonces tendremos que hacer un paso previo de unificación de datos.**

## 2 cualitativo.
Los datos cualitativos tratan con características y descripciones que no se pueden medir fácilmente, pero que se pueden observar subjetivamente, como los olores, los sabores, las texturas, el atractivo y el color. 

Cuando se clasifica o categoriza algo, se crea datos cualitativos o de atributos. Hay tres tipos principales de datos cualitativos.
### 2.1 Los datos binarios 
colocan las cosas en una de dos categorías mutuamente excluyentes: correcto / incorrecto, verdadero / falso o aceptar / rechazar.
### 2.2  datos no ordenados o nominales
asignamos elementos individuales a categorías con nombre que no tienen un valor o rango implícito o natural. Si revisé una caja de manís y grabé el color de cada una en mi hoja de trabajo, serían datos nominales.podría usar el análisis de chi-cuadrado para ver si hay diferencias estadísticamente significativas en las cantidades de cada color en un cuadro.
### 2.3 datos ordenados u ordinales
en los que los elementos se asignan a categorías que tienen algún tipo de orden implícito o natural, como "Corto, Medio o Alto". puede ser la calificacion que le das a un restaurante. Todos coinciden en que este tipo de datos es apropiado para crear gráficos de barras, pero más allá de eso, no se sabe más.

# **Eliminar duplicados y value_counts()**
## value_counts()
(normalize: bool = False, sort: bool = True, ascending: bool = False, bins=None, dropna: bool = True) -> Series

## drop_duplicates(subset="columna", inplace=True)


````python
dic_to_dataf = {'Categoria':['Verduras','Lacteos','Carnes Frias','Panaderia','Panaderia','Panaderia'], 
                'Producto':['Pepino','Leche Entera','Salchichas','Almojábana','Pan con queso','Almojábana'],
                'Precio':[600,2300,4650,1200,1500,1200],
                'Fecha':['2021-01-02','2019-01-31','2020-05-25','2020-08-12','2020-11-03','2020-08-12'],
                "Presentacion":["unidad", "paquete", "bolsa","unidad","unidad","unidad"]}
my_df = pd.DataFrame(dic_to_dataf)

# value_counts permite hacer un conteo o normalizar la suma a 1
print(my_df["Producto"].value_counts(sort=True))
print("-"*70)
print(my_df["Producto"].value_counts(normalize=True))

# podemos eliminar duplicados con drop_duplicates
print("-"*70)
print(my_df.drop_duplicates(subset="Producto", inplace=True)) # como vemos se modifica el archivo original
print(my_df)
print("-"*70)
print(my_df["Producto"].value_counts(sort=True)) # corroboramos que ya no hay elementos repetidos en la columna "Porductos"
````

# **Groupby** 🔗
esto nos permitira agrupar todos los datos que tengan el mismo id, ya seleccionado podremos hacer alguna operacion, a todos los datos. 
por ejemplo vemos que nuestro my_df tiene en categoria 2 muestra iguales (panaderia) podriamos sacar los precios de todas muestras panaderia y calcular la media, o hacer otra operacion. 



````python
print(my_df.groupby("Categoria")["Precio"].mean())
print("-"*70)
print(my_df.groupby("Categoria")["Precio"].agg([min,max, sum]))  
# como vemos el unico ID al que se aplican las operaciones de sumatoria es a panaderia ya que es el unico que esta repetido y se puede agrupar

#supermercado_to_dic.groupby(["Categoria","Presentacion"])["Precio"].agg([min, max, sum])
````

# **Pivot tables** 📄

DataFrame.pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True) 

en conclusion hace lo mismo que group by, pero con diferente sintaxis

````python
# values = valores sobre los cuales se realizara la funcion 
# indice de la columna que se buscara datos repetidos para relaizar la operacion en sus values.
print(my_df.pivot_table(values="Precio", index="Categoria")) # si no se le especifica nada se sacara la media en automatico
print("-"*70)

print(my_df.pivot_table(values="Precio", index="Categoria", columns="Presentacion", fill_value=0, margins=True)) # activa el margen all
````
