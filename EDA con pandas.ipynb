# ¬øQu√© es el EDA?
Eda es la sigla en ingl√©s para **Exploratory Data Analysis** y consiste en una de las primeras tareas que tiene que desempe√±ar el Cient√≠fico de Datos. Es cuando revisamos por primera vez los datos que nos llegan, por ejemplo un archivo CSV que nos entregan y deberemos intentar comprender **‚Äú¬øde qu√© se trata?‚Äù, vislumbrar posibles patrones y reconociendo distribuciones estad√≠sticas que puedan ser √∫tiles en el futuro.**    
OJO!, **lo ideal es que tengamos un objetivo que nos hayan ‚Äúadjuntado‚Äù con los datos**, que indique lo que se quiere conseguir a partir de esos datos. Por ejemplo, nos pasan un excel y nos dicen ‚ÄúQueremos predecir ventas a 30 d√≠as‚Äù, √≥ ‚ÄúClasificar casos malignos/benignos de una enfermedad‚Äù, ‚ÄúQueremos identificar audiencias que van a realizar re-compra de un producto‚Äù, ‚Äúqueremos hacer pron√≥stico de fidelizaci√≥n de clientes/abandonos‚Äù, ‚ÄúQuiero detectar casos de fraude en mi sistema en tiempo real‚Äù.

# EDA deconstruido
Al llegar un archivo, lo primero que deber√≠amos hacer es intentar responder:

1. ¬øCu√°ntos registros hay? ¬øSon demasiado pocos? ¬øSon muchos y no tenemos Capacidad (CPU+RAM) suficiente para procesarlo?
2. ¬øEst√°n todas las filas completas √≥ tenemos campos con valores nulos?

3. ¬øQue datos son discretos y cuales continuos? * tipos de datos 
Muchas veces sirve obtener el tipo de datos: texto, int, double, float
4. tipo de problema: 
  
  4.1 Si es un problema de tipo supervisado:
¬øCu√°l es la columna de ‚Äúsalida‚Äù? ¬øbinaria, multiclase?
¬øEsta balanceado el conjunto salida?

  4.2 ¬øCuales parecen ser features importantes? ¬øCuales podemos descartar?
¬øSiguen alguna distribuci√≥n?
  4.3 ¬øHay correlaci√≥n entre features (caracter√≠sticas)?

  4.4 ¬øEstamos ante un problema dependiente del tiempo? Es decir un TimeSeries.

5. ¬øCuales son los Outliers? (unos pocos datos aislados que difieren dr√°sticamente del resto y ‚Äúcontaminan‚Äù √≥ desv√≠an las distribuciones)
Podemos eliminarlos? es importante conservarlos?
son errores de carga o son reales?
6. ¬øTenemos posible sesgo de datos? (por ejemplo perjudicar a clases minoritarias por no incluirlas y que el modelo de ML discrimine)
Puede ocurrir que tengamos set de datos incompletos y debamos pedir a nuestro cliente/proveedor √≥ interesado que nos brinde mayor informaci√≥n de los campos, que aporte m√°s conocimiento √≥ que corrija campos.

# Tipos de datos Estad√≠sticos: Discretos, continuos, categ√≥ricos y ordinales
[Tipos de datos Estad√≠sticos: Discretos, continuos, categ√≥ricos y ordinales](https://https://www.tecnologias-informacion.com/tipos-estadisticos.html#:~:text=Los%20datos%20discretos%20son%20un,lo%20general%2C%20implica%20n%C3%BAmeros%20enteros.&text=Por%20ejemplo%2C%20podr%C3%ADamos%20realizar%20un,el%20interior%20(datos%20discretos).)

Al trabajar con estad√≠sticas, es importante reconocer los diferentes tipos de datos: num√©ricos (discretos y continuos), categ√≥ricos y ordinales.
La mayor√≠a de los datos caen en uno de dos grupos: num√©ricos o categ√≥ricos.

**En t√©rminos generales, cuando mides algo y le das un valor num√©rico, creas datos cuantitativos. Cuando clasifica o juzga algo, crea datos cualitativos.** 
En el nivel m√°s alto, existen dos tipos de datos: 
## 1 cuantitativo 
La informaci√≥n cuantitativa trata de n√∫meros y cosas que puede medir objetivamente: dimensiones como altura, ancho y longitud. Temperatura y humedad.
Como regla general, los recuentos son discretos y las mediciones son continuas.
### 1.1 datos discretos
Los datos discretos son un conteo que no se puede hacer m√°s preciso. Por lo general, implica n√∫meros enteros. Por ejemplo,el numero de hijos, es un entero, no puedes tener 1.5 hijos. 
### 1.2 Los datos continuos
Los datos continuos, por otro lado, podr√≠an dividirse y reducirse a niveles cada vez m√°s finos. Por ejemplo, puede medir la altura de sus hijos en escalas progresivamente m√°s precisas (metros, cent√≠metros, mil√≠metros y m√°s).

Algunos an√°lisis usan datos cuantitativos continuos y discretos al mismo tiempo.

Por ejemplo, podr√≠amos realizar un an√°lisis de regresi√≥n para ver si el peso de los man√≠s (datos continuos) est√° correlacionado con el n√∫mero de man√≠s en el interior (datos discretos).

**Tambi√©n puede que nos pasen m√∫ltiples fuentes de datos, por ejemplo un csv, un excel y el acceso a una base de datos. Entonces tendremos que hacer un paso previo de unificaci√≥n de datos.**

## 2 cualitativo.
Los datos cualitativos tratan con caracter√≠sticas y descripciones que no se pueden medir f√°cilmente, pero que se pueden observar subjetivamente, como los olores, los sabores, las texturas, el atractivo y el color. 

Cuando se clasifica o categoriza algo, se crea datos cualitativos o de atributos. Hay tres tipos principales de datos cualitativos.
### 2.1 Los datos binarios 
colocan las cosas en una de dos categor√≠as mutuamente excluyentes: correcto / incorrecto, verdadero / falso o aceptar / rechazar.
### 2.2  datos no ordenados o nominales
asignamos elementos individuales a categor√≠as con nombre que no tienen un valor o rango impl√≠cito o natural. Si revis√© una caja de man√≠s y grab√© el color de cada una en mi hoja de trabajo, ser√≠an datos nominales.podr√≠a usar el an√°lisis de chi-cuadrado para ver si hay diferencias estad√≠sticamente significativas en las cantidades de cada color en un cuadro.
### 2.3 datos ordenados u ordinales
en los que los elementos se asignan a categor√≠as que tienen alg√∫n tipo de orden impl√≠cito o natural, como "Corto, Medio o Alto". puede ser la calificacion que le das a un restaurante. Todos coinciden en que este tipo de datos es apropiado para crear gr√°ficos de barras, pero m√°s all√° de eso, no se sabe m√°s.

# **Eliminar duplicados y value_counts()**
## value_counts()
(normalize: bool = False, sort: bool = True, ascending: bool = False, bins=None, dropna: bool = True) -> Series

## drop_duplicates(subset="columna", inplace=True)


````python
dic_to_dataf = {'Categoria':['Verduras','Lacteos','Carnes Frias','Panaderia','Panaderia','Panaderia'], 
                'Producto':['Pepino','Leche Entera','Salchichas','Almoj√°bana','Pan con queso','Almoj√°bana'],
                'Precio':[600,2300,4650,1200,1500,1200],
                'Fecha':['2021-01-02','2019-01-31','2020-05-25','2020-08-12','2020-11-03','2020-08-12'],
                "Presentacion":["unidad", "paquete", "bolsa","unidad","unidad","unidad"]}
my_df = pd.DataFrame(dic_to_dataf)

# value_counts permite hacer un conteo o normalizar la suma a 1
print(my_df["Producto"].value_counts(sort=True))
print("-"*70)
print(my_df["Producto"].value_counts(normalize=True))

# podemos eliminar duplicados con drop_duplicates
print("-"*70)
print(my_df.drop_duplicates(subset="Producto", inplace=True)) # como vemos se modifica el archivo original
print(my_df)
print("-"*70)
print(my_df["Producto"].value_counts(sort=True)) # corroboramos que ya no hay elementos repetidos en la columna "Porductos"
````

# **Groupby** üîó
esto nos permitira agrupar todos los datos que tengan el mismo id, ya seleccionado podremos hacer alguna operacion, a todos los datos. 
por ejemplo vemos que nuestro my_df tiene en categoria 2 muestra iguales (panaderia) podriamos sacar los precios de todas muestras panaderia y calcular la media, o hacer otra operacion. 



````python
print(my_df.groupby("Categoria")["Precio"].mean())
print("-"*70)
print(my_df.groupby("Categoria")["Precio"].agg([min,max, sum]))  
# como vemos el unico ID al que se aplican las operaciones de sumatoria es a panaderia ya que es el unico que esta repetido y se puede agrupar

#supermercado_to_dic.groupby(["Categoria","Presentacion"])["Precio"].agg([min, max, sum])
````

# **Pivot tables** üìÑ

DataFrame.pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True) 

en conclusion hace lo mismo que group by, pero con diferente sintaxis

````python
# values = valores sobre los cuales se realizara la funcion 
# indice de la columna que se buscara datos repetidos para relaizar la operacion en sus values.
print(my_df.pivot_table(values="Precio", index="Categoria")) # si no se le especifica nada se sacara la media en automatico
print("-"*70)

print(my_df.pivot_table(values="Precio", index="Categoria", columns="Presentacion", fill_value=0, margins=True)) # activa el margen all
````

# **Missing values** ‚ùå
Detectar Missing Values:

Eliminar valores faltantes:

Reemplazar valores faltanes:


````python
# podremos ver en True las muestras que no tienen un valor asignado
print(my_df2.isna())
````

````python
# podriamos a√±adir una nueva muestra(producto, fila) con un valor faltante
# es importante agregar el ignore_index ya que si no nos dara error por que el nuevo dato no coincide con las dimenciones del DF
print(my_df)
print("-"*70)
fmy_df = my_df.append({"Categoria":"Verdura", "Producto": "Tomate"}, ignore_index=True)
print(fmy_df) # como vemos no especificamos el precio y fecha de nuestro nuevo producto tomate y aparece como "NaN"
````

````python
# podemos revisar si alguna columna del DF tiene algun dato faltante con any
print(fmy_df.isna().any())
print("-"*70)
print(fmy_df.isna().sum()) # tambien podemos calcular cuanto errores hay usando una funcion 
````

````python
# Eliminar valores faltantes:
print(fmy_df.dropna(inplace=True))# implace indica que lo eliminara 
print(fmy_df)
print("-"*70)

# o Reemplazar valores faltanes:
fmy_df = my_df.append({"Categoria":"Verdura", "Producto": "Tomate"}, ignore_index=True)
print(fmy_df.fillna(0))
````
